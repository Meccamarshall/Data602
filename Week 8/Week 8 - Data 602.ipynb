```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load data and create DataFrame
airbnb_data = df = pd.read_csv('https://raw.githubusercontent.com/Meccamarshall/Data602/main/Week%208/Brooklyn_airbnb.csv')
df = pd.DataFrame(airbnb_data)

# Create scatter plot with Seaborn
sns.scatterplot(data=df, x='longitude', y='latitude', hue='neighbourhood',
                palette='Set2', alpha=0.7)

# Set title and axis labels
plt.title('Airbnb Listings by Neighborhood')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Change size of axis labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Change legend properties
plt.legend(title='Neighborhood', fontsize=10, title_fontsize=12,
           loc='center left', bbox_to_anchor=(1, 0.5))

# Add annotation
plt.annotate('Brooklyn', xy=(-73.95, 40.8), xytext=(-73.6, 40.9),
             arrowprops=dict(arrowstyle='->'), fontsize=12)

# Show plot
plt.show()
```
